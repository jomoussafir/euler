```python
import numpy as np
```

```python
a = 1001
n = int((a-1)/2)

x = np.array([(2*i+1)**2 for i in range(0,n+1)])
dx = np.array([i*2 for i in range(0,n+1)])

print(x)
print(x - dx)
print(x - 2*dx)
print(x - 3*dx)

x.sum() + (x - dx).sum() + (x - 2*dx).sum() + (x - 3*dx).sum() - 3
```
