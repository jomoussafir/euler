```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon

from numpy.linalg import det
import pandas as pd

def zero_in(a):
    sgn_det = list(np.sign([det(np.vstack([a[i], a[((i+1)%3)]])) for i in range(3)]))
    if all([s>0 for s in sgn_det]) or all([s<0 for s in sgn_det]):
        return True
    else:
        return False
```

```python
z = pd.read_csv("p102_triangles.txt", header=None).to_numpy()
nb_triangle = z.shape[0]
z_list = [z[i].reshape((3,2)) for i in range(nb_triangle)]
```

```python
for i in range(20):
    a = np.random.normal(size=6).reshape((3,2))
    if zero_in(a):
        polygon = Polygon(a, alpha=0.2, ec="red", lw=1, fc="w")

        fig, ax = plt.subplots(1,1)

        ax.add_patch(polygon)
        ax.scatter(a[:,0],a[:,1])
        ax.scatter(0,0)
        plt.show()
```
