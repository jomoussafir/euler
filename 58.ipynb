```python
import sympy
```

```python
def pr_sp(k_max):
    l=[1]

    for k in range(1,k_max+1):
        for i in range(4):
            l.append(l[-1]+2*k)
    l_pr = []
    for i in l:
        if sympy.isprime(i):
            l_pr.append(i)

    return (2*k_max+1, len(l_pr)/len(l))

pr_sp(13000)
```

```python
for k_max in range(13000, 14000):
    print(pr_sp(k_max))
```
