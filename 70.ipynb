```python
import numpy as np
np.set_printoptions(linewidth=750)

import pandas as pd
import matplotlib.pyplot as plt

from numba import jit

from sympy import factorint, Mul, primefactors, sqrt, totient, prime

import sys

from tqdm.notebook import tqdm
```

```python
%%time

l = []
for i in tqdm(range(1,10000001)):
    it = totient(i)

    a = list(str(i))
    b = list(str(it))
    
    if set(a) == set(b):
        s = set(a)
        flag = True
        for c in s:
            if a.count(c) != b.count(c):
                flag = False
        if flag:
            l+=[[i, it]]
```

```python
x = np.array(l, dtype=int)
x[:3]
```
