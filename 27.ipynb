```python
import numpy as np

import pandas as pd
from sympy import *
from sympy.ntheory import factorint
from sympy import factorint, Mul, primefactors, sqrt
from sympy import divisors
from sympy.ntheory import isprime
```

```python
z_list = []
for a in range(-1000,1001):
    for b in range(-1000,1001):
#        print("="*10)
        p_list=[]
        n=0
        cont=True
        while cont:
            p_test = n*n + a*n + b
#            print(a,b,n,p_test)
            if isprime(p_test):
                p_list.append(p_test)
            else:
                cont=False
            n+=1
#        print((a,b,len(p_list)))
        z_list.append((a,b,len(p_list)))

z_df = pd.DataFrame(z_list, columns=["a","b","length"])
z_df = z_df.loc[z_df.length>0]
z_df
```
