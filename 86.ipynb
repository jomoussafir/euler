## Pythagorean triplets

```python
%%time
M = 12

@jit
def pythag(M):
    l = []
    for a in np.arange(1,M):
        for b in np.arange(1,a+1):
            if a**2 + b**2 <= M:
                l.append([a*a - b*b, 2*a*b, a*a + b*b])
    return l

z_df = pd.DataFrame(pythag(M), columns=["a","b","c"])
```

```python
%%time
M=20

@jit
def spider(M): 
    cube_set=set()
    
    for a in np.arange(1,M+1):
        for b in np.arange(1,a+1):
            for c in np.arange(1,b+1):
                                
                p_test = np.zeros((6,3))
                l_test = np.zeros((6))
                
                x = a*b/(b+c)
                l = np.sqrt(b**2 + x**2) + np.sqrt((a-x)**2 + c**2)
                l_test[0]=l
                p_test[0,:] = [x,b,c]

                z = b*c/(a+b)
                l = np.sqrt(b**2 + z**2) + np.sqrt((c-z)**2 + a**2)
                l_test[1]=l
                p_test[1,:] = [0,b,z]
                
                y = b*c/(a+c)
                l = np.sqrt(c**2 + y**2) + np.sqrt((b-y)**2 + a**2)
                l_test[2]=l
                p_test[2,:] = [0,y,c]
                
                x = a*c/(b+c)
                l = np.sqrt(c**2 + x**2) + np.sqrt((a-x)**2 + b**2)
                l_test[3]=l
                p_test[3,:] = [x,0,c]
                
                y = a*b/(a+c)
                l = np.sqrt(a**2 + y**2) + np.sqrt((b-y)**2 + c**2)
                l_test[4]=l
                p_test[4,:] = [a,y,0]
                
                z = a*c/(a+b)
                l = np.sqrt(a**2 + z**2) + np.sqrt((c-z)**2 + b**2)
                l_test[5]=l
                p_test[5,:] = [a,0,z]
                
                i = np.argmin(l_test)
                
                if l_test[i] - np.floor(l_test[i]) <= 1e-5:
#                    print("+")
#                    print((a,b,c))
#                    print(p_test[i,:])
#                    print(l_test[i])
                    cube_set.add((a,b,c))
                    
    return cube_set

```
