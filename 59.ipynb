```python
import numpy as np
import pandas as pd
import string, re
alphabet = string.ascii_letters[:26]
import matplotlib.pyplot as plt
```

```python
f = open("0059_cipher.txt")
l = f.readlines()
f.close()
z = [int(i) for i in l[0].split(",")]
```

```python
pwd = ['a','b','c']
key = [ord(s) for s in pwd]*485
[chr(key[i]^z[i]) for i in np.arange(10)]
```

```python
msg_list=[]
score_list=[]
for i in np.arange(len(alphabet)):
    for j in np.arange(len(alphabet)):
        for k in np.arange(len(alphabet)):
            pwd = [alphabet[i], alphabet[j], alphabet[k]]
            key = [ord(s) for s in pwd]*485
            msg = ''.join([chr(key[i]^z[i]) for i in np.arange(len(z))])
            msg_list.append(msg)
            score_list.append(msg.count(" the "))
```

```python
plt.plot(score_list)
```
