```python
import numpy as np
import matplotlib.pyplot as plt

import pandas as pd
np.set_printoptions(linewidth=700)
```

```python
def f(x,y):
    return (x*x - x - y*y, 2*x*y - y - np.pi)

def f_iter(x,n):
    x_iter = np.zeros((n,2))
    x_iter[0,:] = x
    for i in np.arange(1,n):
        x_iter[i,0] = x_iter[i-1,0]*x_iter[i-1,0] - x_iter[i-1,0] - x_iter[i-1,1]*x_iter[i-1,1]
        x_iter[i,1] = 2*x_iter[i-1,0]*x_iter[i-1,1] - x_iter[i-1,1] - np.pi
        
    return x_iter
```

```python
x = (1,0)
plt.plot(f_iter(x, 10).T)
```

```python
a = 20
x = np.linspace(0,a,a+1)
nb_x = len(x)

y = np.linspace(-a,a,50)
nb_y = len(y)

z = np.meshgrid(x,y)
print(nb_x, nb_y)
print(x)
print(y)
z[0].shape

w = [None]*2
w[0] = np.zeros((nb_y, nb_x))
w[1] = np.zeros((nb_y, nb_x))

for i in np.arange(nb_x):
    w[0][:,i], w[1][:,i] = f(z[0][:,i], z[1][:,i])

plt.subplot(2,1,1)
for i in np.arange(nb_x):
    plt.plot(z[0][:,i], z[1][:,i], '-')

plt.subplot(2,1,2)
for i in np.arange(nb_x):
    plt.plot(w[0][:,i], w[1][:,i], '-')
```
